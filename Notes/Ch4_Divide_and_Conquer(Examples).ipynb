## 1. Divide and Conquer Algorithms

The divide and conquer style algorithms *divide* the problem to be solved into 2 or more sub-problems, *conquer* each of the problems separately, and then *combine* the solutions if needed after each sub-problem was solved on their own.

**Divide** the problem into one or more subproblems that are smaller instances of the same problem.

**Conquer** the subproblems by solving them recursively

**Combine** the subproblem solutions to form a solution to the original problem.

## 2. Computing the max or min value of a list

Consider the problem of computing the maximum or minimum value of a list. To solve this problem via a divide and conquer algorithm we can use the following observation.

The minimum value of a list $A = \left[a_0, a_1, a_2, \ldots , a_{n-1}\right]$ is the least value between the minimum of the left half of the list $A[0:n/2]$ and the minimum of the right half of the list $A[n/2:n]$

Another way to view this is either the minimum value is on the left half of the list $A$ or it is on the right half of the list $A$. This means we can divide the work between finding the min on the left half and finding the min on the right half. After we find the minimum of each half, we can compare them to see which is the lowest for the entire list. The steps are summarized in the points below.

1. Split the list into two parts left and right,
2. find the minimum of each half of the list separetely,
3. fompare the two minima to see which is the minimum for the entire list.

These steps are implemented using recursion in the divide and conquer algorithm below. The base case of the recursion is a list with one value, in which case the one value is the maximum and the minimum (since there is only 1 value).

```python
def min_DnC(A,p,r):
    """finds the minimum value of A[p:r] using a divide and conquer approach"""

    if p==r:          # Base Case
        return A[r]

    ######################################## divide
    q = (p+r)//2      # split point

    ######################################## conquer
    l_min = min_DnC(A,p,q)      # left min
    r_min = min_DnC(A,q+1,r)    # right min

    ######################################## combine
    if l_min < r_min:
        return l_min
    else:
        return r_min
```

```python
import random

n = 10

# basic random list
A = [random.randrange(99) for i in range(n)]
print(A)

min_DnC(A,0,n-1)
```

```python
# right edge case
A = [random.randrange(99) for i in range(n)]+[-10]
print(A)

min_DnC(A,0,n)
```

```python
# left edge case
A = [-10] + [random.randrange(99) for i in range(n)]
print(A)

min_DnC(A,0,n)
```

## 3. Computing the inner product of vectors

Many linear algebra problems have standard divide and conquer solutions which are based on the same principle as the algorithm above. When one has a list of values we are working with, it is (perhaps) better to split the list somehow, process each half, then combine the solutions. When we combine this observation (based on memory usage) with recursion (a programming technique), we have a powerful framework to use for building efficient algorithms.

Consider computing the inner product of two vectors.
$$
\begin{align*}
\vec{x} &= \left[x_0, x_1, x_2, \ldots, x_{n-1}\right] \\
\vec{y} &= \left[y_0, y_1, y_2, \ldots, y_{n-1}\right] \\
\end{align*}
$$

The *inner product* of $\vec{x}$ and $\vec{y}$ is the sum of products of corresponding entries from $\vec{x}$ and $\vec{y}$,
$$
\vec{x}^T\vec{y} = x_0y_0 + x_1y_1 + x_2y_2 + \ldots + x_{n-1}y_{n-1}.
$$

We can represent this in summation notation as follows,
$$
\vec{x}^T\vec{y} = \sum_{i=0}^{n-1} x_iy_i.
$$

The key observation combining the memory operations of partial lists and recursion is to *divide* the summation loop into separate parts, *conquer* the summation on each part, and *combine* the results after each was computed separately.
$$
\vec{x}^T\vec{y} = \sum_{i=0}^{n-1} x_iy_i  =  \sum_{i=0}^{q-1} x_iy_i  +  \sum_{i=q}^{n-1} x_iy_i.
$$
Applying recursion we see the following relation,

$$
inner\_prod\_DnC(x,y,n) = inner\_prod\_DnC(x_{left},y_{left},q) + inner\_prod\_DnC(x_{right},y_{right},n-q).
$$

The recurrence relation above is what is implemented below.

```python
def inner_prod_DnC(x,y,n):
    """returns the inner-product of vectors x and b"""
    
    ######## Base Cases
    #if n==0: return 0
    if n==1: return x[0]*y[0]
    
    ####################################### divide
    q = n//2
    
    ####################################### conquer
    left = inner_prod_DnC(x[:q],y[:q],q)
    right = inner_prod_DnC(x[q:],y[q:],n-q)
    
    ####################################### combine
    return left + right
```

```python
n = 8

x = [random.randint(-9,9) for i in range(n)]
print(x)
y = [random.randint(-9,9) for i in range(n)]
print(y)

output = inner_prod_DnC(x,y,n)
print("x.y =",output)
```

```python
import numpy as np

def check_inner(x,y,output):
    """checks the inner product with numpy"""
    
    TOL = 10**(-12)
    
    np_x = np.array(x)
    np_y = np.array(y)

    diff = np_x.T @ np_y - np.array([[output]])
    if np.math.fabs(diff) < TOL:
        return True
    else:
        return False
```

```python
if check_inner(x,y,output):
    print("correct!")
```

## 4. Vector Addition

As another example we consider vector addition. Given two vectors,
$$
\begin{align*}
\vec{x} &= \left[x_0, x_1, x_2, \ldots, x_{n-1}\right] \\
\vec{y} &= \left[y_0, y_1, y_2, \ldots, y_{n-1}\right] \\
\end{align*}
$$

The *vector sum* $\vec{z} = \vec{x} + \vec{y}$ is the vector whose entries are the sum of corresponding entries from $\vec{x}$ and $\vec{y}$,
$$
\vec{x} + \vec{y} = \vec{z} = \left[ x_0+y_0 , x_1+y_1 , x_2+y_2 , \ldots , x_{n-1}+y_{n-1}\right].
$$
i.e. $z_i = x_i + y_i$ for $i=0,1,2,\ldots,n-1$.

To apply the divide and conquer technique here we *divide* the work of computing the sum of corresponding entries and then *conquer* each by computing the sums of entries for just the left half and the right half separately. In this case, we copy the sums for each half back to the original desired memory locations.



**Note**

It is not always required that the algorithm copy the data to work on each half and then copy the results. In some cases, this is easier to code or required for different reasons, in other cases this is not desired and an *in-place* algorithm is preferred which works only using the original memory that was given plus a constant ammount of additional memory.

```python
def vector_add_DnC(x,y,z,n):
    """computes the sum of vectors x and b"""
    
    ######## Base Cases
    if n==1:
        z[0] = x[0]+y[0]
        return
    
    ####################################### divide
    q = n//2
    
    x_left  = [x[i] for i in range(q)]
    x_right = [x[i] for i in range(q,n)]
    y_left  = [y[i] for i in range(q)]
    y_right = [y[i] for i in range(q,n)]
    z_left  = [0 for i in range(q)]
    z_right = [0 for i in range(q,n)]

    
    ####################################### conquer
    vector_add_DnC(x_left  , y_left  , z_left  , q)
    vector_add_DnC(x_right , y_right , z_right , n-q)

    ####################################### combine
    for i in range(q):
        z[i] = z_left[i]
    for i in range(q,n):
        z[i] = z_right[i-q]
```

```python
n = 8

x = [random.randint(-9,9) for i in range(n)]
print(x)
y = [random.randint(-9,9) for i in range(n)]
print(y)
z = [0 for i in range(n)]

vector_add_DnC(x,y,z,n)
print(z)
```

```python
def check_sum(x,y,z):
    """checks the sum with numpy"""
    
    TOL = 10**(-12)
    
    np_x = np.array(x)
    np_y = np.array(y)

    diff = (np_x + np_y) - np.array(z)
    if np.linalg.norm(diff) < TOL:
        return True
    else:
        return False
```

```python
if check_sum(x,y,z):
    print("correct!")
```
