### a. Rod-Cutting Algorithms

````python
def cut_rod(p,n):
    """recursive rod-cutting program"""
    #print(n)
    if n==0:
        return 0
    q = -float("inf")
    for i in range(1,n+1): # this loop must start at 1 or the recursion won't stop
        q = max( q, p[i] + cut_rod(p,n-i) )
    return q
````

````python
def memoizes_cut_rod(p,n):
    """non-recursive starting code"""
    
    r = [-float("inf") for i in range(n+1)]
    
    return memoized_cut_rod_aux(p,n,r)

def memoized_cut_rod_aux(p,n,r):
    """recursive code"""
    if r[n]>=0:         # check the memo table
        return r[n]
    if n==0:            # base case
        q = 0
    else:               # recursively compute sub-problem
        q = -float("inf")
        for i in range(1,n+1):
            q = max( q , p[i] + memoized_cut_rod_aux(p,n-i,r) )
    
    r[n] = q            # save memo to table
    
    return q
````

### 3. Re-construct an LCS from the completed table c and the original sequences X and Y

This is problem 14.4-2 from the CLRS textbook.

````python
def LCS_Length(X,Y,m,n):
    """compute a longest common sub-sequence of the sequences X and Y of lenghts m and n"""
    c = [[0 for j in range(n+1)] for i in range(m+1) ] # (rows) 0...m   and  (cols) 0...n
    print("dimensions:",(len(c),len(c[0])))
    
    for i in range(1,m+1):
        for j in range(1,n+1):
            if X[i]==Y[j]:                # match
                c[i][j] = c[i-1][j-1] + 1
            elif c[i-1][j] >= c[i][j-1]:  # check for max of previous two sizes
                c[i][j] = c[i-1][j]
            else:
                c[i][j] = c[i][j-1]
    return c
````

````python
def re_construct_LCS(c,X,Y,m,n):
    """re-construct the LCS solution from the table c"""
    # start at last position
    i = m-1
    j = n-1
    size = c[i][j]
    S = [None for k in range(size)] # this should store the sub-sequence
    ##########################################
    # insert your code here
    k = size - 1

    while i > 0 and j > 0:
        if X[i] == Y[j]:
            S[k] = X[i]
            k = k - 1
            i = i - 1
            j = j - 1

        elif c[i - 1][j] >= c[i][j - 1]:
            i = i - 1
        else:
            j = j - 1
    
    
    
    
    ##########################################
    return S
````

### 4. Give a memo-ized version of LCS-LENGTH that runs in O(mn) time.

This is 14.4-3 from the CLRS textbook.

````python
def LCS_Length_Memoized(X,Y,m,n):
    """compute a longest common sub-sequence of the sequences X and Y of lenghts m and n"""
    
    c = [[None for j in range(n+1)] for i in range(m+1) ] # (rows) 0...m   and  (cols) 0...n
    #print("dimensions:",(len(c),len(c[0])))

    i_start = m
    j_start = n
    LCS_Length_Memoized_Recursive(X,Y,m,n,c,i_start,j_start)
    return c

def LCS_Length_Memoized_Recursive(X, Y, m, n, c, i, j):
    """recursive code for memo-ization of the LCS-Length Algorithm"""    
    #################################################
    if c[i][j] is not None:
        return c[i][j]
    
    if i == 0 or j == 0:
        c[i][j] = 0
    
    elif X[i] == Y[j]:
        c[i][j] = LCS_Length_Memoized_Recursive(X, Y, m, n, c, i - 1, j - 1) + 1
        LCS_Length_Memoized_Recursive(X, Y, m, n, c, i - 1, j)      
        LCS_Length_Memoized_Recursive(X, Y, m, n, c, i, j - 1)    
    
    else:
        up = LCS_Length_Memoized_Recursive(X, Y, m, n, c, i - 1, j)
        left = LCS_Length_Memoized_Recursive(X, Y, m, n, c, i, j - 1)
        c[i][j] = max(up, left)
    #################################################
    return c[i][j]
````

### 5. Implement an $\mathcal{O}(n^2)$ time algorithm to find the longest monotonically increasing subsequences of a sequence of $n$ numbers.

This is exercise 14.4-5 from the CLRS textbook

````python
def LCMIS_Length(X, m):
    """compute a longest common monotonically increasing sub-sequence of the sequence X of length m"""
    
    Y = sorted(set(X))  
    n = len(Y)

    
    c = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if X[i - 1] == Y[j - 1]:
                temp = c[i - 1][j - 1]
                c[i][j] = temp + 1
            else:
                c[i][j] = max(c[i - 1][j], c[i][j - 1])

    
    S = []
    i, j = m, n
    while i > 0 and j > 0:
        if X[i - 1] == Y[j - 1]:
            S.append(X[i - 1])
            i -= 1
            j -= 1
        elif c[i - 1][j] >= c[i][j - 1]:
            i -= 1
        else:
            j -= 1
    S.reverse()
  
    return c, S
````
