### a. Activity-Selection

```python
class activity:
    """these are the activites to schedule"""

    def __init__(self,name="",start = 0,finish = 0):
        """construct an activity to schedule"""
        self.name = "a_"+str(name)
        self.s = start
        self.f = finish

    def __str__(self):
        """printing utililty"""
        out = str(self.name) + "\t\t|\ttime slot: [" + str(self.s) + "," + str(self.f) + ")"
        return out
```

```python
import random

def generate_activity_set(n):
    """generates a set of activities to use for an Activity-Selection problem of size n"""

    # time parameters (rules for scheduling)
    earliest_start = 8    # this stands for 8 AM
    latest_stop = 22      # this stands for 10 PM (12 + 10 = 22 hours after midnight)
    max_time = 3          # this is the max reservation time
    
    # generate a set of "random" activites to schedule
    S = []
    for i in range(n):
        name = i          # a_0, a_1, etc.
        
        # determine a random starting time
        start = random.randrange(earliest_start,latest_stop)

        # determine a random stop time
        finish = random.randrange(start + 1,start + max_time + 1)
        while finish > latest_stop:
            finish = random.randrange(start + 1,start + max_time + 1)

        # create the activity and append to the list
        S.append(activity( name , start , finish ))
    return S
```

### b. Brute Force for Standard Activity-Selection Problem

```python
def feasible(S):
    """verify if the set of (sorted) activities are feasible"""
    for i in range(len(S)-1):
        a_i = S[i]   # current
        a_j = S[i+1] # next
        
        if a_j.s < a_i.f: # a_j (next one) starts before a_i finishes
            return False
    
    return True
```

```python
import itertools

def Brute_Force_Activity_Selection(S):
    """Brute Force Solution to the Activity-Selection problem for list of Activities"""
    sorted_S = sorted(S,key=lambda a:a.f) # sort list
    n = len(sorted_S)  # number of items
    
    max_size = 0       # size of best sub-set
    max_comb = None    # save the best combination
    count = 0
    
    for i in range(1,n+1):
        combs = itertools.combinations(sorted_S,i) # compute selections of size i
    
        # loop over the iterator of combinations
        for comb in combs:
            count += 1
            if feasible(comb): # check if the combination is feasible
                if i > max_size:
                    max_size = i
                    max_comb = comb
    
    print("total combinations checked:",count)
    return max_size, max_comb
```

### e. {0,1}-Knapsack Problem codes

```python
class item:
    """these are the items to steal"""

    def __init__(self,name="",size=1,value=1):
        """construct an item to steal"""
        self.name = "x_"+str(name)
        self.size = size
        self.value = value

    def __str__(self):
        """printing utililty"""
        out = str(self.name) + "  |  (" + str(self.size) + "," + str(self.value) + ")"
        return out
```

```python
def generate_item_set_simple(N,M):
    """generates a set of items to use for a Knapsack problem of size N,M"""

    # generate a set of "random" items to "steal"
    Items = []
    value = 20
    size = 1
    for i in range(N):
        if N < 27:
            name = chr(ord('A')+i) # for small problems use letters x_A, x_B, etc.
        else:
            name = str(i)          # for large problems use numbers x_0, x_1, etc.
        
        size = random.randrange(size+1,size+3)   # random increasing sizes
        value = random.randrange(value-5,value)  # random decreasing values
        
        Items.append(item( name , size , value )) # create item and append
    return Items
```

```python
def Brute_Force_Knapsack(Items,M):
    """Brute Force Solution to the Knapsack problem for list of Items and capacity M"""
    N = len(Items) # number of items total
    
    max_score = 0      # initialize (overall) max score
    max_size = 0       # size of best sub-set
    max_comb = None    # save the best combination
    
    
    for i in range(1,N+1):
        # generate the n choose i combinations
        combs = itertools.combinations(Items,i)
    
        # loop over the iterator of combinations
        for comb in combs:
            score = 0                       # initialize score (for sub-set)
            size = 0
            for j in range(i):              # loop over items of the sub-set
                item_obj = comb[j]          # extract item
                score += item_obj.value     # add value to score (for sub-set)
                size += item_obj.size       # add size to total
            
            if score > max_score and size <= M:
                max_score = score
                max_size = size
                max_comb = [ items for items in comb]
    
    ################################################################## output below
    return max_score, max_size, max_comb
```

```python
def recursive_selection_Memoization(S):
    """solve the activity selection problem recursively"""
    count = 0
    n = len(S)
    c = [[None for j in range(n+2)] for i in range(n+2)] # memo table
    #################################################################
    # insert your additional memo tables and storage here
    additional_memo = []

    for i in range(n+2):
        temp_row = []

        for j in range(n+2):
            temp_row.append(None)

        additional_memo.append(temp_row)
    #################################################################
    
    
    # nested memoization code
    def memoize(sorted_S,i,j):
        """perform memoization as a wrapper to code below"""
        if c[i][j] == None:
            c[i][j] = recursive_selectionMemo(sorted_S,i,j) # compute (if needed)
        return c[i][j]          # return lookup
    
    # nested recursive code
    def recursive_selectionMemo(sorted_S,i,j):
        """generic recursive solution to the Activity-Selection Problem"""
        nonlocal count
        count += 1
        
        max_val = -float("inf")              # lowest value
        ai = sorted_S[i]     # extract activity i
        aj = sorted_S[j]     # extract activity j
        
        for k in range(1,len(sorted_S)-1):   # loop over other activites
            ak = sorted_S[k] # extract activity k
            if ai.f <= ak.s and ak.f <= aj.s:
                # ak is compatible with ai and aj
                value = memoize(sorted_S,i,k) + memoize(sorted_S,k,j) + 1
                if value > max_val: # check for a larger total
                    max_val = value
                    additional_memo[i][j] = k
        if max_val > 0:
            return max_val
        else:
            #print("BASE CASE")
            return 0
    
    
    ############################################# driver code #########################################
    sorted_S = [activity("_",0,0)] + sorted(S,key=lambda a:a.f) + [activity("_",24,24)] # sort list and add dummy entries
    N = len(sorted_S) # new length
    
    output = recursive_selectionMemo(sorted_S,0,N-1)

    # re-construct the solution from the best array
    #############################################################
    # insert your code here
    def helper(i,j):
        k = additional_memo[i][j]

        if k is not None:
            helper(i,k)
            solution.append(sorted_S[k])
            helper(k,j)


    solution = []
    helper(0,N-1)
    
    #############################################################
    print("total recursive calls:",count)
    return output,solution
```

### 3. Room Scheduling

This is exercise 15.1-4 in the CLRS textbook.

```python
def room_scheduler(S):
    """greedy algorithm for scheduling rooms """
    ################################################################

    solution_list = []
    # insert your code here
    sorted_S = sorted(S, key=lambda a: a.s)
    for activity in sorted_S:
        placed = False
        for room in solution_list:      
            if room[-1].f <= activity.s:
                room.append(activity)
                placed = True
                break
        if not placed:
            solution_list.append([activity])
    
    
    
    
    
    ################################################################
    return solution_list
```

### 5. Simplified {0,1}-Knapsack

This is exercise 15.2-3 in the CLRS textbook.

```python
def knapsack_simple(Items,M):
    """Solve the Knapsack problem for the set of Items and total capacity M (using Dynamic Programming)"""
    ################################################################
    # insert your code here
    solution = []
     
    
    n = len(Items)
    
   
    K = [[0 for w in range(M+1)] for i in range(n+1)]
    
    for i in range(1, n+1):
        item = Items[i-1]
        for w in range(M+1):
            if item.size <= w:
                K[i][w] = max(K[i-1][w], K[i-1][w - item.size] + item.value)
            else:
                K[i][w] = K[i-1][w]
    
   
    for i in range(n, 0, -1):
        if K[i][w] != K[i-1][w]:
            solution.append(Items[i-1])
            w -= Items[i-1].size
    
    solution.reverse()
    
    
    
    
    
    
    
    
    ######### return the output of the re-construction method
    return solution
```
