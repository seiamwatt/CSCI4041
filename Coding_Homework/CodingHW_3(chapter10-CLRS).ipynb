````python
class stack:
    """Array(Python-list)-based Stack -Chapter 10 - CLRS"""

    def __init__(self):
        """constructor for stack objects"""
        self.top = -1
        self.size = 11
        self.S = [None for i in range(self.size)] # "array" with initial size 11

    def stack_empty(self):
        """returns True if the stack is emtpy and False otherwise"""
        if self.top == -1:
            return True
        else:
            return False
    
    def push(self,x):
        """adds x to the stack"""
        if self.top == self.size:
            print("overflow")
            return
        else:
            self.top = self.top + 1
            self.S[self.top] = x

    def pop(self):
        """returns the last element added to the stack"""
        if self.stack_empty():
            print("underflow")
            return
        else:
            self.top = self.top-1
            return self.S[self.top+1]

    def __repr__(self):
        """string printing utility for python print"""
        return str(self.S)
````

````python
class queue:
    """Array(Python-list)-based Queue -Chapter 10 - CLRS"""

    def __init__(self):
        """constructor for queue objects"""
        self.head = 0                    # most recently added item
        self.tail = self.head            # first open spot
        self.size = 11
        self.Q = [None for i in range(self.size)] # "array" with initial size 11

    def queue_empty(self):
        """returns True if the queue is empty and False otherwise"""
        if self.tail == self.head:
            return True
        else:
            return False
    
    def enqueue(self,x):
        """adds x to the queue"""
        if self.head==self.tail+1 or (self.head==0 and self.tail==self.size):
            print("overflow")
            return
        else:
            self.Q[self.tail] = x         # add to "back" of the list
            if self.tail==self.size-1:
                self.tail = 0             # reset for "circular" array access
            else:
                self.tail = self.tail + 1 # move right

    def dequeue(self):
        """returns the last element added to the stack"""
        if self.queue_empty():
            print("underflow")
            return
        else:
            x = self.Q[self.head] # find "oldest" element in queue
            if self.head==self.size-1:
                self.head = 0
            else:
                self.head = self.head + 1
            return x

    def __repr__(self):
        """string printing utility for python print"""
        return str(self.Q)
````

````python
class linkedlist:
    """a python implementation of singly linked list from ch10 -CLRS"""

    ##################################################################### nested sub-class for nodes
    class node:
        """(nested) node class for linkedlist"""
        def __init__(self,k=None,n=None):
            """constructor for node objects"""
            self.next = n
            self.key = k
    #####################################################################

    
    def __init__(self):
        """constructor for the linkedlist object"""
        self.head = None
    
    def prepend(self,k):
        """adds (node) x to the front of the list"""
        x = self.node(k)      # create new node with key k
        x.next = self.head    # new node points to old head
        self.head = x         # new head is new node
    
    def search(self,k):
        """finds element with key k (returning the node)"""
        x = self.head
        
        while x and x.key!=k:
            x = x.next
        return x
    
    def insert(self,k,k2):
        """insert new key k immediately following old key k2"""
        x = self.node(k)       # new node with key k
        y = self.search(k2)    # node y with key k2

        if y:
            x.next = y.next        # update links
            y.next = x
        else:
            print("node with key =",k2," not found")

    def delete(self,k):
        """deletes the node with key k (if it exists)"""
        x_prev = None
        x = self.head
        
        while x and x.key!=k:
            x_prev = x
            x = x.next
        
        if x:
            if x_prev:
                x_prev.next = x.next
            else:
                self.head = x.next

    def __repr__(self):
        """produces a string of the Linked List"""
        cursor = self.head
        out = "["
        while cursor:
            out += str(cursor.key) + ", "
            cursor = cursor.next
        return out[:-2] + "]"if len(out)>2 else "[]"
````

````python
class deque:
    """Array(Python-list)-based Double-Ended Queue -Chapter 10 - CLRS"""

    def __init__(self):
        """constructor for queue objects"""
        self.size = 11
        self.Q = [None for i in range(self.size)] # "array" with initial size 11
        ###########################################################
        # insert your code here for additional constructor items
        self.count = 0
        self.head = 0
        self.tail = 0 
    
        
        
        
        
        pass

    def queue_empty(self):
        """returns True if the queue is empty and False otherwise"""
        ############################################
        # insert your code here
        if self.count == 0:
            return True
        
        return False
    
    def enqueue(self,x):
        """adds x to the back of the queue"""
        ############################################
        # insert your code here
        
        if self.count == self.size:
            return
        
        self.Q[self.tail] = x
        self.tail = (self.tail + 1) % self.size
        self.count += 1

    def dequeue(self):
        """returns the oldest element from the front of the queue"""
        ############################################
        # insert your code here
        if self.queue_empty():
            return
        
        x = self.Q[self.head]
        self.head = (self.head + 1) % self.size 
        self.count -= 1
        return x


    
    
    def add_first(self,x):
        """adds x to the front of the queue (non-typical)"""
        ############################################
        # insert your code here
        if self.count == self.size:
            return
        

        self.head = (self.head - 1) % self.size
        self.Q[self.head] = x
        self.count += 1



    def remove_last(self):
        """returns the oldest element in the queue (non-typical)"""
        ############################################
        # insert your code here
        if self.queue_empty():
            return

        self.tail = (self.tail - 1) % self.size
        x = self.Q[self.tail]
        self.count -= 1
        
        return x




    def __repr__(self):
        """string printing utility for python print"""
        return str(self.Q)
````
